```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, acf, pacf
import statsmodels.api as sm
from pmdarima import auto_arima
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
warnings.filterwarnings('ignore')
```

```python
# Download Tesla and Ford data from Yahoo Finance
tesla = yf.download('TSLA', start='2013-06-24', end='2024-06-24', interval='1mo')
ford = yf.download('F', start='2013-06-24', end='2024-06-24', interval='1mo')

# Only keep 'Close' prices
tesla = tesla[['Close']].rename(columns={'Close': 'Tesla'})
ford = ford[['Close']].rename(columns={'Close': 'Ford'})

# Merge the two dataframes
df_monthly = pd.concat([tesla, ford], axis=1)
```

```python
# Plot Tesla and Ford prices
plt.figure(figsize=(12, 6))
plt.plot(df_monthly['Tesla'], label='Tesla')
plt.plot(df_monthly['Ford'], label='Ford')
plt.title('Tesla vs Ford Monthly Close Prices')
plt.legend()
plt.show()
```

```python
# Decompose both time series
decompose_tesla = seasonal_decompose(df_monthly['Tesla'], model='additive', period=12)
decompose_ford = seasonal_decompose(df_monthly['Ford'], model='additive', period=12)

# Plot decomposition
decompose_tesla.plot()
plt.show()

decompose_ford.plot()
plt.show()
```

```python
# ADF Test for stationarity
def adf_test(series):
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] > 0.05:
        print("Series is not stationary")
    else:
        print("Series is stationary")

# Test for both Tesla and Ford
adf_test(df_monthly['Tesla'])
adf_test(df_monthly['Ford'])
```

```python
# Differencing if necessary
df_monthly['Tesla_diff'] = df_monthly['Tesla'].diff().dropna()
df_monthly['Ford_diff'] = df_monthly['Ford'].diff().dropna()

# Check again after differencing
adf_test(df_monthly['Tesla_diff'])
adf_test(df_monthly['Ford_diff'])
```

```python
# Train-test split
train_size = len(df_monthly) - 12  # Use the last 12 months as the test set

train_data = df_monthly[:train_size]  # Training data
test_data = df_monthly[train_size:]   # Test data (last 12 months)
```

```python
from pmdarima import auto_arima

# Train SARIMA on training data
sarima_tesla = auto_arima(train_data['Tesla'], seasonal=True, trace=True)
sarima_ford = auto_arima(train_data['Ford'], seasonal=True, trace=True)

# Fit SARIMA for both Tesla and Ford
sarima_model_tesla = sm.tsa.SARIMAX(train_data['Tesla'], order=sarima_tesla.order, seasonal_order=sarima_tesla.seasonal_order).fit()
sarima_model_ford = sm.tsa.SARIMAX(train_data['Ford'], order=sarima_ford.order, seasonal_order=sarima_ford.seasonal_order).fit()

# Forecast the next 12 months for both Tesla and Ford
sarima_forecast_tesla = sarima_model_tesla.forecast(steps=12)
sarima_forecast_ford = sarima_model_ford.forecast(steps=12)
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Prepare data for LSTM
def prepare_data_for_lstm(series, time_steps=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    series_scaled = scaler.fit_transform(series.values.reshape(-1, 1))

    X, y = [], []
    for i in range(time_steps, len(series_scaled)):
        X.append(series_scaled[i-time_steps:i, 0])
        y.append(series_scaled[i, 0])

    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # Reshape for LSTM

    return X, y, scaler

# Prepare the training data for LSTM
X_train_tesla, y_train_tesla, scaler_tesla = prepare_data_for_lstm(train_data['Tesla'])
X_train_ford, y_train_ford, scaler_ford = prepare_data_for_lstm(train_data['Ford'])

# Build the LSTM model
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50))
    model.add(Dense(1))  # Output layer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train the LSTM model for Tesla
model_tesla = build_lstm_model((X_train_tesla.shape[1], 1))
model_tesla.fit(X_train_tesla, y_train_tesla, epochs=20, batch_size=32)

# Train the LSTM model for Ford
model_ford = build_lstm_model((X_train_ford.shape[1], 1))
model_ford.fit(X_train_ford, y_train_ford, epochs=20, batch_size=32)

# Forecast the next 12 months for both Tesla and Ford
def forecast_lstm(model, X, scaler, steps=12):
    predictions = []
    current_input = X[-1]

    for _ in range(steps):
        prediction = model.predict(current_input.reshape(1, X.shape[1], 1))
        predictions.append(prediction[0, 0])
        current_input = np.append(current_input[1:], prediction)

    predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))
    return predictions

forecast_lstm_tesla = forecast_lstm(model_tesla, X_train_tesla, scaler_tesla)
forecast_lstm_ford = forecast_lstm(model_ford, X_train_ford, scaler_ford)
```

```python
from sklearn.metrics import mean_absolute_percentage_error

# Function to calculate MAPE (Mean Absolute Percentage Error)
def calculate_mape(actual, predicted):
    return mean_absolute_percentage_error(actual, predicted) * 100

# Compare SARIMA forecasts with actual test data
print("SARIMA Tesla MAPE:", calculate_mape(test_data['Tesla'], sarima_forecast_tesla))
print("SARIMA Ford MAPE:", calculate_mape(test_data['Ford'], sarima_forecast_ford))

# Compare LSTM forecasts with actual test data
print("LSTM Tesla MAPE:", calculate_mape(test_data['Tesla'].values, forecast_lstm_tesla))
print("LSTM Ford MAPE:", calculate_mape(test_data['Ford'].values, forecast_lstm_ford))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Train Exponential Smoothing for Tesla
exp_smooth_tesla = ExponentialSmoothing(train_data['Tesla'], seasonal='add', seasonal_periods=12).fit()
forecast_exp_tesla = exp_smooth_tesla.forecast(12)

# Train Exponential Smoothing for Ford
exp_smooth_ford = ExponentialSmoothing(train_data['Ford'], seasonal='add', seasonal_periods=12).fit()
forecast_exp_ford = exp_smooth_ford.forecast(12)
```

```python
# Function to ensure forecasts are 1D Series or arrays and lengths match
def ensure_1d_and_correct_length(forecast, actual_length):
    forecast = np.array(forecast).flatten()  # Ensure it's 1D
    if len(forecast) != actual_length:
        raise ValueError(f"Forecast length {len(forecast)} does not match actual data length {actual_length}")
    return forecast

# Function to create DataFrame with actual, forecasted, and residuals for SARIMA, LSTM, and Exponential Smoothing
def create_forecast_dataframe(actual, forecast_sarima, forecast_lstm, forecast_exp, stock_name):
    actual_length = len(actual)
    
    # Ensure all forecasts are 1D and match the length of actual data
    forecast_sarima = ensure_1d_and_correct_length(forecast_sarima, actual_length)
    forecast_lstm = ensure_1d_and_correct_length(forecast_lstm, actual_length)
    forecast_exp = ensure_1d_and_correct_length(forecast_exp, actual_length)
    
    # Create DataFrame
    df = pd.DataFrame({
        f'{stock_name}_Actual': actual.values,
        f'{stock_name}_SARIMA_Forecast': forecast_sarima,
        f'{stock_name}_LSTM_Forecast': forecast_lstm,
        f'{stock_name}_Exp_Smoothing_Forecast': forecast_exp,
        f'{stock_name}_SARIMA_Residuals': actual.values - forecast_sarima,
        f'{stock_name}_LSTM_Residuals': actual.values - forecast_lstm,
        f'{stock_name}_Exp_Smoothing_Residuals': actual.values - forecast_exp
    }, index=actual.index)
    
    return df

# Create DataFrames for Tesla and Ford, including Exponential Smoothing
tesla_df = create_forecast_dataframe(test_data['Tesla'], sarima_forecast_tesla, forecast_lstm_tesla, forecast_exp_tesla, 'Tesla')
ford_df = create_forecast_dataframe(test_data['Ford'], sarima_forecast_ford, forecast_lstm_ford, forecast_exp_ford, 'Ford')

# Combine both DataFrames into one
final_df = pd.concat([tesla_df, ford_df], axis=1)

# Export to Excel
output_file = 'forecasted_values_residuals_with_exp_smoothing.xlsx'
final_df.to_excel(output_file)

print(f"Data successfully exported to {output_file}")
```

```python
from scipy.stats import shapiro

# Function to perform the Shapiro-Wilk test
def perform_shapiro_test(data, label):
    stat, p_value = shapiro(data)
    print(f"\nShapiro-Wilk Test for {label}:")
    print(f"Test Statistic: {stat}")
    print(f"P-value: {p_value}")
    
    # Interpretation of the result
    if p_value > 0.05:
        print(f"{label} appears to be normally distributed (fail to reject H0).")
    else:
        print(f"{label} does not appear to be normally distributed (reject H0).")

# Perform the Shapiro-Wilk test for Tesla's residuals from the SARIMA model
perform_shapiro_test(test_data['Tesla'] - sarima_forecast_tesla, "Tesla SARIMA Residuals")

# Perform the Shapiro-Wilk test for Ford's residuals from the SARIMA model
perform_shapiro_test(test_data['Ford'] - sarima_forecast_ford, "Ford SARIMA Residuals")

```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from pmdarima import auto_arima
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Assuming df_monthly contains the monthly stock prices for Tesla and Ford, and it is already loaded.

# 1. SARIMA Model Forecasting (Auto ARIMA)
def forecast_sarima(series, steps=12):
    # Fit SARIMA using auto_arima
    sarima_model = auto_arima(series, seasonal=True, m=12, trace=True)
    return sarima_model.predict(n_periods=steps)

# 2. Exponential Smoothing Forecasting
def forecast_exponential_smoothing(series, steps=12):
    # Fit Exponential Smoothing model
    exp_model = ExponentialSmoothing(series, seasonal='add', seasonal_periods=12).fit()
    return exp_model.forecast(steps)

# 3. LSTM Model Forecasting
def forecast_lstm(series, steps=12, time_steps=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(series.values.reshape(-1, 1))
    
    # Prepare data for LSTM
    X = []
    for i in range(time_steps, len(scaled_data)):
        X.append(scaled_data[i-time_steps:i, 0])
    
    X = np.array(X)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # Reshape for LSTM
    
    # Build LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
    model.add(LSTM(units=50))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Fit the model
    model.fit(X, scaled_data[time_steps:], epochs=20, batch_size=32, verbose=0)
    
    # Forecast the next 'steps' months
    predictions = []
    current_input = scaled_data[-time_steps:]
    
    for _ in range(steps):
        prediction = model.predict(current_input.reshape(1, time_steps, 1))
        predictions.append(prediction[0, 0])
        current_input = np.append(current_input[1:], prediction)
    
    # Inverse scale the predictions
    return scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

# Forecast for next 12 months using each model for Tesla and Ford
# Tesla Forecasts
sarima_forecast_tesla = forecast_sarima(df_monthly['Tesla'], steps=12)
exp_forecast_tesla = forecast_exponential_smoothing(df_monthly['Tesla'], steps=12)
lstm_forecast_tesla = forecast_lstm(df_monthly['Tesla'], steps=12)

# Ford Forecasts
sarima_forecast_ford = forecast_sarima(df_monthly['Ford'], steps=12)
exp_forecast_ford = forecast_exponential_smoothing(df_monthly['Ford'], steps=12)
lstm_forecast_ford = forecast_lstm(df_monthly['Ford'], steps=12)

# Create DataFrame with the results
forecast_table = pd.DataFrame({
    'Date': pd.date_range(start=df_monthly.index[-1], periods=13, freq='M')[1:],  # Next 12 months
    'Tesla_SARIMA_Forecast': sarima_forecast_tesla,
    'Tesla_LSTM_Forecast': lstm_forecast_tesla.flatten(),
    'Tesla_Exp_Smoothing_Forecast': exp_forecast_tesla,
    'Ford_SARIMA_Forecast': sarima_forecast_ford,
    'Ford_LSTM_Forecast': lstm_forecast_ford.flatten(),
    'Ford_Exp_Smoothing_Forecast': exp_forecast_ford
})

# Display the forecast table
print(forecast_table)
```
